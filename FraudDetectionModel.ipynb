Importing Dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#loading Dataset
card_data=pd.read_csv('creditcard.csv')
```

This dataset is highly unbalanced So we need to Handle the Unbalanced



0-->Normal Transaction

1-->Fradulent Transaction

```python
#Seperating the data for analysis
normal=card_data[card_data.Class==0]
fraud=card_data[card_data.Class==1]
```

```python
#Print the shape of both transactions
print(normal.shape)
print(fraud.shape)
```

```python
#Statical measures of normal transaction data
normal.Amount.describe()
```

```python
#Statical measures of normal transaction data
fraud.Amount.describe()
```

```python
#Comparing Value for bith the Transactions
card_data.groupby('Class').mean()
```

We will Deal with Unbalanced Data using Undersampling.

We will build a sample dataset containing similar distribution of Normal transaction and Fraud Transaction.


Total no of Fraud Transaction is 492 which we got from

 print(fraud.shape)-->(492, 31)


```python
#Taking random Sample of Entire dataset
normal_sample=normal.sample(n=492)
```

Concatenating two dataframes

1st is Normal and 2nd is the Fraud Dataframe

```python
#If we put axis=1 then the values will be added column wise but for axis=0 the values will be added below the normal_sample
new_dataset=pd.concat([normal_sample,fraud],axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

This Step helps to understand if we have good Sample or Bad Sample.


If it was Bad Sample we would have got the Mean Values of the random Selected values and The will be having very high deviation


But in our case the mean values doesn't actually deviate much so it means it is a Good Sample.

```python
new_dataset.groupby('Class').mean()
```

Now we will Split the data into Features and Target and feed into Machine learning Model.

```python
X=new_dataset.drop(columns='Class', axis=1)
Y=new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

Now we will Split these features and labels into Training data and Testing data


We will take the Features present in X and Y and split it into training data and Testing Data So that it is splitted randomly and the corrosponding label will be chosen from the Y

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Training our Machine Learning Model using Logistic Regression

```python
model=LogisticRegression( )
```

Training the model using the Logistic Regression Model with training data.

```python
model.fit(X_train, Y_train)
```

Model Evaluation and giving it Accuracy Score.

```python
X_train_prediction=model.predict(X_train)
training_data_accuracy=accuracy_score(X_train_prediction, Y_train)
```

```python
print("Accuracy on Training data: ",training_data_accuracy)
```

```python
#Accuracy on Test Data
X_test_prediction=model.predict(X_test)
test_data_accuracy=accuracy_score(X_test_prediction, Y_test)
```

```python
print("Accuracy Score on Test Data: ",test_data_accuracy)
```
